1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python
import os
archivo = open('Emisiones_CO2.csv', 'r') #la r de read, es para que solo lo abra en modo lectura, encoding = 'ANSI')
```

```python
# Leer primeras lineas para ver que tomo bien el archivo (se ve tambien si hay 'enies' o acentos)
linea= archivo.readline()
contador=0
while contador < 5:
    print(linea)
    linea= archivo.readline()
    contador+=1
```

```python
#Creo diccionario para guardar los datos
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python
#Carga los datos en el diccionario
archivo.seek(0) #retorna el marcado al inicio, es decir se se usa denuevo read le desde el comienzo y no desde donde se quedo
next(archivo) #saltea la primer linea que estan los titulos
for linea in archivo:
    linea = linea.strip() # Elimina los espacios y saltos de linea al principio y al final de las cadenas
    campos = linea.split('|') # Devuelve una lista separada por lo que le pida
    dicc_emisiones['cod_pais'].append(campos[0])
    dicc_emisiones['nom_pais'].append(campos[1])
    dicc_emisiones['region'].append(campos[2])
    dicc_emisiones['anio'].append(campos[3])
    dicc_emisiones['co2'].append(campos[4])
    dicc_emisiones['co2_percapita'].append(campos[5])
```

```python
archivo.close() #cierro el archivo para que no me consuma memoria RAM
```

```python
## d) 
nulos_codPais = 0
for elemento in dicc_emisiones['cod_pais']:
    if elemento == "": #Para este caso los elementos vacios no son tomados como none
        nulos_codPais += 1
print('La cantidad de nulos para cod_pais es: ',nulos_codPais)
 
nulos_nomPais = 0
for elemento in dicc_emisiones['nom_pais']:
    if elemento == "":
        nulos_nomPais += 1
print('La cantidad de nulos para nom pais es: ',nulos_nomPais)
 
nulos_anio = 0
for elemento in dicc_emisiones['anio']:
    if elemento == "":
        nulos_anio += 1
print('La cantidad de nulos para anio es: ',nulos_anio)
 
nulos_co2 = 0
for elemento in dicc_emisiones['co2']:
    if elemento == "":
        nulos_co2 += 1
print('La cantidad de nulos para co2 es: ',nulos_co2)
 
nulos_co2Percapita = 0
for elemento in dicc_emisiones['co2_percapita']:
    if elemento == "":
        nulos_co2Percapita += 1
print('La cantidad de nulos para co2_percápita es: ',nulos_co2Percapita)



nulos_totales =  nulos_codPais + nulos_nomPais + nulos_anio + nulos_co2 + nulos_co2Percapita
print('La cantidad de nulos totales es de: ',nulos_totales)

```
